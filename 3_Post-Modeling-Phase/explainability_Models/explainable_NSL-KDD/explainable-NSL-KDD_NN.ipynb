## Compute Permutation Importance

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.base import BaseEstimator, ClassifierMixin
from sklearn.inspection import permutation_importance


# Define a wrapper for the Keras model to make it Scikit-learn compatible
class KerasWrapper(BaseEstimator, ClassifierMixin):
    def __init__(self, model):
        self.model = model

    def fit(self, X, y):
        pass  # We don't need to re-train the model

    def predict(self, X):
        return (self.model.predict(X) > 0.5).astype("int32")  # Convert to binary classification

    def predict_proba(self, X):
        proba = self.model.predict(X)
        return np.hstack([1 - proba, proba])  # Format for Scikit-learn (two columns)

wrapped_model = KerasWrapper(model)
    
# Compute permutation importance
result = permutation_importance(
    wrapped_model, 
    X_test_scaled, 
    y_test, 
    scoring='accuracy', 
    n_repeats=10, 
    random_state=42
)

# Extract feature importances
importances = result.importances_mean
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(12, 8))
plt.barh(range(len(importances)), importances[indices], color="skyblue")
plt.yticks(range(len(importances)), [selected_features[i] for i in indices])
plt.gca().invert_yaxis()
plt.title("Feature Importances (Permutation Importance)")
plt.show()
```

## Morris Sensitivity Analysis

```python
from interpret.blackbox import MorrisSensitivity
from interpret import show
import pandas as pd

# Define a wrapper for the predict function to return numpy arrays
def predict_fn(data):
    # Ensure predictions are converted to NumPy arrays
    predictions = model.predict(data)
    return predictions.flatten()  # Flatten to ensure a 1D array

# Initialize Morris Sensitivity Analyzer
msa = MorrisSensitivity(
    model=predict_fn,  # Pass the wrapper as the model
    data=X_test_scaled,
    feature_names=selected_features
)

# Generate global explanations
explanation = msa.explain_global()

# Extract Morris sensitivity scores
sensitivity_df = pd.DataFrame(explanation.data()["scores"], index=selected_features)

# Sort the sensitivity scores for better interpretability
sorted_sensitivity_df = sensitivity_df.sort_values(by=sensitivity_df.columns[0], ascending=False)
print(sorted_sensitivity_df)

# Visualize explanations
show(explanation, max_features=len(selected_features))
```

## LOFO IMPORTANCE

```python
from lofo import LOFOImportance, Dataset, plot_importance
from sklearn.model_selection import KFold
import pandas as pd

def calculate_lofo_importance(df, target_column, feature_names, cv_splits=4, scoring="roc_auc"):
    """
    Calculate LOFO (Leave One Feature Out) importance using the specified validation scheme and scorer.

    Parameters:
    - df: pandas.DataFrame containing the dataset.
    - target_column: str, the name of the target column in the dataset.
    - feature_names: list, names of features to evaluate for LOFO importance.
    - cv_splits: int, number of cross-validation splits.
    - scoring: str, the scoring metric to use (e.g., "roc_auc", "accuracy").

    Returns:
    - importance: pandas.DataFrame containing LOFO importances with mean and std values.
    """
    # Define cross-validation strategy
    cv = KFold(n_splits=cv_splits, shuffle=False, random_state=None)  # Avoid shuffling for time-series compatibility

    # Define the dataset for LOFO analysis
    dataset = Dataset(df=df, target=target_column, features=selected_features)

    # Initialize LOFOImportance object
    lofo_imp = LOFOImportance(dataset, cv=cv, scoring=scoring)

    # Compute feature importances
    importance = lofo_imp.get_importance()

    # Plot the feature importances
    plot_importance(importance, figsize=(12, 20))

    return importance

# Example usage
%time importance = calculate_lofo_importance(df=df_test, target_column="binary_attack", feature_names=selected_features)

# Display the importance DataFrame
print(importance)
```

## SHAP LEQ and GEU

```python
import numpy as np
from scipy.stats import spearmanr

# Functions for Metrics
def calculate_local_explanation_quality(predicted_explanations, actual_model_outputs):
    """
    Calculate the Local Explanation Quality (Fidelity Local).

    Args:
        predicted_explanations (np.ndarray): Predicted explanation outputs (e.g., SHAP predicted values).
        actual_model_outputs (np.ndarray): Actual local model outputs.

    Returns:
        float: Local explanation quality score.
    """
    N = len(predicted_explanations)
    fidelity_local = 1 - (np.sum((predicted_explanations - actual_model_outputs) ** 2) / N)
    return fidelity_local

def calculate_global_explanation_utility(feature_importance_values, domain_importance_values):
    """
    Calculate the Global Explanation Utility using Spearman's Rank Correlation.

    Args:
        feature_importance_values (np.ndarray): Feature importance values (e.g., SHAP feature importance).
        domain_importance_values (np.ndarray): Domain knowledge importance values.

    Returns:
        float: Spearman's Rank Correlation.
    """
    if len(feature_importance_values) != len(domain_importance_values):
        raise ValueError("The number of domain importance scores must match the number of SHAP features.")
    correlation, _ = spearmanr(feature_importance_values, domain_importance_values)
    return correlation
```

# 2. SHAP Global Explanation Utility

```python
from scipy.stats import spearmanr

def calculate_global_explanation_utility(shap_feature_importance, domain_importance_scores):
    """
    Calculate the global explanation utility by comparing SHAP feature importance
    with domain knowledge importance scores using Spearman's Rank Correlation.

    Args:
        shap_feature_importance (numpy array): SHAP feature importance scores (absolute mean SHAP values).
        domain_importance_scores (numpy array): Domain knowledge feature importance scores (normalized).

    Returns:
        float: Spearman's Rank Correlation coefficient as the global explanation utility.
    """
    if len(shap_feature_importance) != len(domain_importance_scores):
        raise ValueError("The number of SHAP feature importances must match the number of domain importance scores.")
    
    # Calculate Spearman's Rank Correlation
    correlation, _ = spearmanr(shap_feature_importance, domain_importance_scores)
    return correlation

# SHAP Feature Importance: Average absolute SHAP values
shap_feature_importance = np.mean(np.abs(shap_values.values), axis=0)

domain_importance_scores = np.array([
    0.07,  # level
    0.10,  # src_bytes (important for intrusion detection)
    0.08,  # dst_host_srv_count
    0.07,  # dst_bytes
    0.08,  # count
    0.06,  # protocol_type
    0.07,  # dst_host_diff_srv_rate
    0.07,  # dst_host_same_srv_rate
    0.05,  # service
    0.06,  # rerror_rate
    0.04,  # logged_in
    0.07,  # dst_host_srv_serror_rate
    0.07,  # serror_rate
    0.06   # dst_host_rerror_rate
])
# Normalize domain importance scores to sum to 1
domain_importance_scores /= domain_importance_scores.sum()

# Ensure lengths match
if len(shap_feature_importance) != len(domain_importance_scores):
    raise ValueError("The number of domain importance scores must match the number of SHAP features.")

# Calculate SHAP Global Explanation Utility
global_utility = calculate_global_explanation_utility(shap_feature_importance, domain_importance_scores)

# Outputs
print(f"SHAP Global Explanation Utility (Spearman's Rank Correlation): {global_utility:.4f}")
```

## LIME EXPLAINABILITY

```python
from lime.lime_tabular import LimeTabularExplainer
import numpy as np
import pandas as pd

# Ensure your data is prepared
X_train_array = np.array(X_train_scaled)  # Training data as a NumPy array
X_test_array = np.array(X_test_scaled)    # Test data as a NumPy array

# Initialize the LimeTabularExplainer
lime_explainer = LimeTabularExplainer(
    training_data=X_train_array,         # Training data
    feature_names=selected_features,         # Feature names
    class_names=['malicious', 'normal'], # Class labels for your model
    mode='classification',               # Set mode to 'classification'
    discretize_continuous=True           # Discretize continuous features
)

# Define the modified prediction function
def predict_proba_lime(X):
    X_df = pd.DataFrame(X, columns=selected_features)  # Convert to DataFrame for consistency
    probabilities = model.predict(X_df)  # Get probabilities for the positive class
    return np.column_stack((1 - probabilities, probabilities))  # Add probabilities for both classes

# Pick a test instance to explain
test_instance = X_test_array[0]  # First test instance

# Explain the prediction for the selected test instance
lime_explanation = lime_explainer.explain_instance(
    data_row=test_instance,          # Test instance to explain
    predict_fn=predict_proba_lime,   # Prediction function (returns probabilities)
    num_features=14                  # Limit explanation to top 10 features
)

# Show the explanation in a human-readable format
lime_explanation.show_in_notebook(show_table=True)

# Visualize LIME explanation as a chart
lime_explanation.as_pyplot_figure()
```

## 1. LIME Local Explanations

```python
from scipy.stats import spearmanr
import numpy as np

# Extract LIME feature importance
lime_feature_importance = np.zeros(len(selected_features))
for feature, weight in lime_explanation.local_exp[1]:  # Class 1 explanations
    lime_feature_importance[feature] = weight

# Normalize LIME feature importance for global utility
lime_feature_importance_normalized = np.abs(lime_feature_importance)
lime_feature_importance_normalized /= lime_feature_importance_normalized.sum()

# Get actual model predictions
actual_model_outputs = model.predict(X_test_scaled).flatten()  # Use probabilities for class 1
lime_predicted_explanations = model.predict(X_test_scaled).flatten()  # Use the same model outputs for simplicity

# Domain Importance Scores (same as before)
domain_importance_scores = np.array([
    0.07,  # level
    0.10,  # src_bytes (important for intrusion detection)
    0.08,  # dst_host_srv_count
    0.07,  # dst_bytes
    0.08,  # count
    0.06,  # protocol_type
    0.07,  # dst_host_diff_srv_rate
    0.07,  # dst_host_same_srv_rate
    0.05,  # service
    0.06,  # rerror_rate
    0.04,  # logged_in
    0.07,  # dst_host_srv_serror_rate
    0.07,  # serror_rate
    0.06   # dst_host_rerror_rate
])

# Normalize Domain Importance Scores
domain_importance_scores /= domain_importance_scores.sum()
```
