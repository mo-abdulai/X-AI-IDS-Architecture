# Exploratory Data Analysis

## IMPORT LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from warnings import filterwarnings

filterwarnings("ignore")
%matplotlib inline

pd.set_option("display.float_format", lambda x: "%.3f" % x)
plt.rcParams["figure.figsize"] = (10, 6)
```

# Load the Dataset

```python
# Load the dataset
file_path = '../../Datasets/CIC-IDS-2017/Friday-WorkingHours-Afternoon-PortScan.pcap_ISCX.csv'
df = pd.read_csv(file_path)

# Display the first few rows
print(df.head())

# Display dataset dimensions
print(f"Dataset contains {df.shape[0]} rows and {df.shape[1]} columns.")
```

# Dataset Overview

### (a) Summary Information

```python
# Display column data types and memory usage
print(df.info())

# Display summary statistics for numerical columns
print(df.describe().T)
```

### (b) Missing Values

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing values per column:\n", missing_values)
```

### (c) Dataset Insights

```python
duplicates = df.duplicated().sum()
print(f"Dataset contains {duplicates} duplicate rows.")
```

# 4. Target Variable Analysis

```python
# Unique values and counts in the target column
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
print(df['label'].value_counts())


# Visualize target distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='label', palette='Set2')
plt.title("Target Variable Distribution")
plt.xticks(rotation=45)
plt.show()
```

## 5. Numeric Features Analysis

### (a) Basic Distribution

```python
non_finite_columns = df[numeric_columns].columns[(~np.isfinite(df[numeric_columns])).any()]
print(f"Columns with non-finite values: {non_finite_columns}")
```

```python
# Replace inf and -inf with NaN
df[['flow_bytes/s', 'flow_packets/s']] = df[['flow_bytes/s', 'flow_packets/s']].replace([np.inf, -np.inf], np.nan)

# Option A: Replace NaN with the column mean
for col in ['flow_bytes/s', 'flow_packets/s']:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
# Check again for non-finite values
non_finite_columns = df[['flow_bytes/s', 'flow_packets/s']].columns[(~np.isfinite(df[['flow_bytes/s', 'flow_packets/s']])).any()]
print(f"Columns with non-finite values after replacement: {non_finite_columns}")
```

```python
# Select numeric columns
numeric_columns = df.select_dtypes(include=["int64", "float64"]).columns

# Plot histograms
df[numeric_columns].hist(figsize=(16, 12), bins=20, color='skyblue')
plt.tight_layout()
plt.show()
```

## (b) Boxplots for Outliers

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

for col in ['flow_duration', 'total_fwd_packets', 'total_backward_packets']:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=df[col], color="lightblue")
    plt.title(f"Boxplot of {col}")
    plt.show()
```

## (c) Correlation Analysis

```python
# Compute and visualize correlation matrix
plt.figure(figsize=(12, 8))
correlation_matrix = df[numeric_columns].corr()
sns.heatmap(correlation_matrix, cmap="coolwarm", annot=False)
plt.title("Correlation Matrix")
plt.show()
```

# 6. Key Feature Exploration

### (a) Relationship Between Features and Target

```python
# print(df.columns)
# df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Example: Flow Duration vs Target
plt.figure(figsize=(8, 5))
sns.kdeplot(data=df, x='flow_duration', hue='label', fill=True, common_norm=False, palette="crest")
plt.title("Flow Duration Distribution by Target")
plt.show()
```

### (b) Average Packet Size vs Target

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(data=df, x='average_packet_size', hue='label', fill=True, common_norm=False, palette="coolwarm")
plt.title("Average Packet Size by Target")
plt.show()
```

### (c) Fwd Packets/s and Bwd Packets/s

```python
# Scatterplot of Forward and Backward Packets per Second
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='fwd_packets/s', y='bwd_packets/s', hue='label', alpha=0.6)
plt.title("Fwd Packets/s vs Bwd Packets/s by Target")
plt.show()
```

# 7. Categorical Flags Analysis

### a) Flag Counts

```python
# Sum of flag columns grouped by target
flag_columns = ['fin_flag_count', 'syn_flag_count', 'rst_flag_count', 'psh_flag_count', 'ack_flag_count']
flag_summary = df[flag_columns + ['label']].groupby('label').sum()
print(flag_summary)

# Visualize flag counts
flag_summary.T.plot(kind='bar', figsize=(12, 6), stacked=True)
plt.title("Flag Counts by Target")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

## (b) URG and CWE Flags

```python
# URG and CWE Flag Analysis
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='urg_flag_count', hue='label', palette='Set1')
plt.title("URG Flag Count by Target")
plt.show()

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='cwe_flag_count', hue='label', palette='Set2')
plt.title("CWE Flag Count by Target")
plt.show()
```

# 8. Advanced Analysis

## (a) Protocol Features

```python
if 'flow_bytes/s' in df.columns:
    plt.figure(figsize=(8, 5))
    sns.kdeplot(data=df, x='flow_bytes/s', hue='label', fill=True, common_norm=False, palette="crest")
    plt.title("Flow Bytes/s Distribution by Target")
    plt.show()
```

### (b) Subflow Packet Analysis

```python
# Subflow packets vs target
plt.figure(figsize=(8, 5))
sns.kdeplot(data=df, x='subflow_fwd_packets', hue='label', fill=True, common_norm=False, palette="viridis")
plt.title("Subflow Fwd Packets by Target")
plt.show()

plt.figure(figsize=(8, 5))
sns.kdeplot(data=df, x='subflow_bwd_packets', hue='label', fill=True, common_norm=False, palette="magma")
plt.title("Subflow Bwd Packets by Target")
plt.show()
```
